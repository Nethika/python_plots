```python
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import pandas as pd
import numpy as np
```

```python
iplot(cf.datagen.lines().iplot(asFigure=True,
                               kind='scatter',xTitle='Dates',yTitle='Returns',title='Returns'))
```

```python
df = cf.datagen.lines()

iplot([{
    'x': df.index,
    'y': df[col],
    'name': col
}  for col in df.columns], filename='cufflinks/simple-line')
```

```python
#df.iplot(kind='scatter', filename='cufflinks/cf-simple-line') #instead use offline:
iplot(df.iplot(asFigure=True,kind='scatter',filename='cufflinks/cf-simple-line'))
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
iplot(df.scatter_matrix(asFigure=True,filename='cufflinks/scatter-matrix', world_readable=True))
```

```python
df[0:10]
```

```python
df = pd.DataFrame({'a': np.random.randn(1000) + 1,
                   'b': np.random.randn(1000),
                   'c': np.random.randn(1000) - 1})

iplot(df.iplot(asFigure= True, kind='histogram', filename='cufflinks/basic-histogram'))
```

```python

iplot(df.iplot(asFigure= True, kind='histogram', barmode='stack', bins=100, histnorm='probability', filename='cufflinks/customized-histogram'))
```

```python
iplot(df.iplot(asFigure=True,kind='histogram', subplots=True, shape=(3, 1), filename='cufflinks/histogram-subplots'))
```

```python
iplot(df.iplot(asFigure= True, kind='spread', filename='cufflinks/basic-histogram'))
```

```python
iplot(df.iplot(asFigure= True, kind='bubble', x='a',y='b',size='c'))
```

```python
iplot(df.iplot(asFigure= True, kind='surface', color='rdylbu'))
```

```python
iplot(df.iplot(asFigure= True, kind='bar'))
```

```python
iplot(df.iplot(asFigure= True, kind='box'))
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/widgets/master/ipython-examples/311_150k.csv', parse_dates=True, index_col=1)
df.head(10)
```

```python
series = df['Complaint Type'].value_counts()[:20]
series.head(10)
```

```python
iplot(series.iplot(asFigure= True,kind='bar', yTitle='Number of Complaints', title='NYC 311 Complaints',
             filename='cufflinks/categorical-bar-chart'))
```
